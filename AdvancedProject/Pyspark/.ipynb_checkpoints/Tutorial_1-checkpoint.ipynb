```python
from pyspark import SparkContext, SQLContext
```

```python
import pandas as pd
```

```python
df_pd = pd.read_csv('imdb.csv')
```

```python
df_pd.head()
```

```python
sc = SparkContext().getOrCreate()
s = SQLContext(sc)
```

```python
# sc.stop()
```

```python
df_spark = s.read.options(header=True).csv('imdb.csv')
```

```python
df_spark.show(5)
```

```python
df_spark.select(['Title','Rank']).show(5)
```

```python
df_spark.columns #same: pandas
```

```python
new_df_pd = df_pd[['Title','Rank']]
```

```python
new_df_pd.head(2)
```

```python
new_df_spark = df_spark.select(['Title','Rank'])
```

```python
new_df_spark.show(2)
```

```python
new_df_pd.columns = ['a','b']
print(new_df_pd.head(2))
new_df_pd.rename(columns={'a':'Title','b':'Rank'}, inplace=True)
print(new_df_pd.head(2))
```

```python
new_df_spark.show(2)
new_df_spark_1 = new_df_spark.toDF('a','b')
new_df_spark_1.show(2)
new_df_spark_1_renamed = new_df_spark_1.withColumnRenamed('a','Title') #multiple columns not possible
new_df_spark_1_renamed.show(2)
```

```python
new_df_spark.withColumnRenamed('Title','a')
new_df_spark.show(2)
```

```python
import pyspark.sql.functions as F # similar to numpy
```

```python
new_df_spark_3 = new_df_spark.withColumn('log(Rank)', F.log(new_df_spark['Rank']))
new_df_spark_3.show(2)
```

```python
df_spark.groupBy("Rating").count().show()
```

```python
df_spark.select(F.countDistinct("Rating")).show()
```

```python
df_spark_1 = df_spark.withColumn('cond', F.when(df_spark['Rank']>8, 'Very Good')\
                   .when(df_spark['Rank'].between(5,8), 'Good')\
                   .otherwise('Poor'))
```

```python
df_spark_1.select(['Title','cond']).show(10)
```

```python
df_spark_1[df_spark_1['cond']=='Very Good']['Title','cond'].show()
```

```python
df_spark = df_spark.withColumnRenamed('Revenue (Millions)','Revenue')
df_spark.columns
```

```python
# change datatype of revenue
```

```python
from pyspark.sql.types import DoubleType, FloatType, IntegerType
```

```python
df_spark = df_spark.withColumn('Revenue', df_spark['Revenue'].cast('float'))
df_spark = df_spark.withColumn('Rank', df_spark['Rank'].cast('int'))
```

```python
df_spark.dtypes
```

```python
df_spark.select(['Rank']).show(5)
# df_spark['Rank'] : use like this only to filter
```

```python
func = F.udf(lambda x:x+1, IntegerType()) # User defined function
df_spark = df_spark.withColumn('Rank', func(df_spark['Rank']))
```

```python
df_spark.select('Rank').show(5)
```

```python
df_spark.describe('Rank').show() # specific columns
df_spark.describe().show() # all columns
```

```python
# Percentile(IQR)
df_spark.selectExpr('percentile_approx(rank, array(0.25,0.5,0.75)) as Rank').show()
```

```python
df_spark.createOrReplaceTempView('sql_table') # temporary table lazy initializer
df_spark_sql_table = s.sql("select * from sql_table where Rating like '7%'")
```

```python
df_spark_sql_table.select(['Title','Rating']).show()
```
