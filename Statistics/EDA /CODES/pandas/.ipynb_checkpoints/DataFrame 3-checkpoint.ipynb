```python
bond.columns = [col_name.replace(' ','_') for col_name in bond.columns]
bond.columns
```

```python
def convert_123(number):
    return str(number)+' millions'

columns = ['Box Office','Budget','Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(convert_123)

bond.head()
```

```python
def rating(row):
    actor = row[1]
    budget = row[4]
    
    if actor=='Sean Connery':
        return 'Somewhat good!'
    elif actor=='Daniel Craig' and budget>100:
        return 'awesome!'
    else:
        return 'Don\'t know about this one!'

bond.apply(rating, axis='columns') #axis=1
```
