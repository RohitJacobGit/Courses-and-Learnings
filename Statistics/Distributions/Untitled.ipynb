Bivariate

#Assumptions
#cov(x,x) = 1 or SD along x-axis is 1
#cov(y,y) = 1 or SD along y-axis is 1
#corr(x,y) = 0

```python
mean = np.array([0.0,0.0]) # Assume mean along x-axis and y-axis as 0
cov_mat = np.array([[1,0],[0,1]])
dist = np.random.multivariate_normal(mean,cov_mat,500)
#dist.shape 500*2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
plt.hist(dist[:,0])
plt.hist(dist[:,1])
```

```python
mean = np.array([0.0,0.0]) # Assume mean along x-axis and y-axis as 0
cov_mat = np.array([[1,0.5],[0.5,1]]) #changed the corr b/n x and y
dist = np.random.multivariate_normal(mean,cov_mat,500)
#dist.shape 500*2
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```
