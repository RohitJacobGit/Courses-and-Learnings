```python
pd.read_csv('worldstats.csv')
world = pd.read_csv('worldstats.csv', index_col=['country','year'])
```

```python
s = world.stack() #series
s.head(3)
#row size doubled and halved the columns
```

```python
s.unstack().unstack().unstack()
```

```python
s.unstack(0)
s.unstack(1)
s.unstack(2)
s.unstack(-1)
s.unstack(-2)
s.unstack(-3)
```

```python
s.unstack(level=[0,1])
s.unstack(level=['country','year'])
```

```python
s.unstack('year',fill_value=0)
```

```python
sales = pd.read_csv('salesmen.csv', parse_dates=['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head(3)
```

```python
sales['Salesman'].value_counts()
```

```python
# use pivot to reduce rows only when you have less classes in categorical column
sales.pivot(index='Date',columns='Salesman',values='Revenue')
```

```python
food = pd.read_csv('foods.csv')
food.head(3)
```

```python
food.pivot_table(index='Gender',values='Spend',aggfunc='sum') #group by gender
food.pivot_table(index='City',values='Spend',aggfunc='mean') #group by city
food.pivot_table(index='Item',values='Spend',aggfunc='sum') #group by item
food.pivot_table(index=['Gender','Item'],values='Spend',aggfunc='sum') #group by item
```
