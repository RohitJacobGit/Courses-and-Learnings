```python
#http://fac.ksu.edu.sa/sites/default/files/business-statistics-for-contemporary-decision-making-by-ken-black_0.pdf
#Pg-147
# According to the U.S. Census Bureau, approximately 6% of all workers in Jackson,
# Mississippi, are unemployed. In conducting a random telephone survey in Jackson,
# what is the probability of getting two or fewer unemployed workers in a sample of 20?
binom.cdf(k=2,n=20,p=0.06)
```

```python
from scipy.stats import poisson
```

```python
#Pg-156
# Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2
# customers every 4 minutes. What is the probability of exactly 5 customers arriving in a
# 4-minute interval on a weekday afternoon? The lambda for this problem is 3.2 customers
# per 4 minutes. The value of x is 5 customers per 4 minutes. The probability of 5 customers
# randomly arriving during a 4-minute interval when the long-run average has been 3.2
# customers per 4-minute interval is
poisson.pmf(5,3.2)
```

```python
#Pg-156
# Bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4 minutes. What is the probability of having more than 7 customers in
# a 4-minute interval on a weekday afternoon?
1-poisson.cdf(7,3.2)
```

```python
#pg-157
# A bank has an average random arrival rate of 3.2 customers every 4 minutes. What
# is the probability of getting exactly 10 customers during an 8-minute interval?
poisson.pmf(10,6.4)
```

```python
# Pg-166
# Suppose 18 major computer companies operate in the United States and that 12 are
# located in California’s Silicon Valley. If three computer companies are selected randomly from the entire list, what is the probability that one or more of the selected
# companies are located in the Silicon Valley?
from scipy.stats import hypergeom
hypergeom.sf(0,18,3,12) #(x-1,N,n,A)
```

```python
# A western city has 18 police officers eligible for promotion. Eleven of the 18 are
# Hispanic. Suppose only five of the police officers are chosen for promotion and that
# one is Hispanic. If the officers chosen for promotion had been selected by chance
# alone, what is the probability that one or fewer of the five promoted officers would
# have been Hispanic? What might this result indicate?
hypergeom.cdf(1,18,5,11)
```
